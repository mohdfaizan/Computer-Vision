```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
%matplotlib inline
```

```python
dark_horse = cv2.imread('C:/Users/Mohd Faizan Umar/Downloads/Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/horse.jpg')
show_horse = cv2.cvtColor(dark_horse,cv2.COLOR_BGR2RGB)

dark_rainbow = cv2.imread('C:/Users/Mohd Faizan Umar/Downloads/Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/rainbow.jpg')
show_rainbow = cv2.cvtColor(dark_rainbow,cv2.COLOR_BGR2RGB)

dark_bricks = cv2.imread('C:/Users/Mohd Faizan Umar/Downloads/Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/bricks.jpg')
show_bricks = cv2.cvtColor(dark_bricks,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(show_horse)
```

```python
# for openCV the colour channel order is BGR
hist_values = cv2.calcHist([dark_bricks],channels=[0],mask=None,histSize=[256],ranges=[0,256]) # here 0 represents Blue channel
```

```python
hist_values.shape
```

```python
plt.plot(hist_values)
```

```python
hist_values1 = cv2.calcHist([dark_horse],channels=[0],mask=None,histSize=[256],ranges=[0,256]) # here 0 represents Blue channel
plt.plot(hist_values1)
```

```python
img = dark_bricks
colors = ('b','g','r')
for i,color in enumerate(colors):
    histr = cv2.calcHist([img],channels=[i],mask=None,histSize=[256],ranges=[0,256])
    plt.plot(histr,color)
```

```python
gorilla = cv2.imread('C:/Users/Mohd Faizan Umar/Downloads/Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/gorilla.jpg',0)

def display(img,cmap=None):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap=cmap)
```

```python
display(gorilla)
```

```python
hist_value = cv2.calcHist([gorilla],channels=[0],mask=None,histSize=[256],ranges=[0,256])
plt.plot(hist_value)
```

```python
eq_gorilla = cv2.equalizeHist(gorilla)
display(eq_gorilla,cmap='gray')
```

```python
hist_value = cv2.calcHist([eq_gorilla],channels=[0],mask=None,histSize=[256],ranges=[0,256])
plt.plot(hist_value)
```
